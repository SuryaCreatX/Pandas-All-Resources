## Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import numpy as np

DataF = {'first' : ['Surya', 'Paul', 'Kalai', 'Varsha', 'NA', 'Praveen', np.nan, np.nan],
        'last' : ['S', 'B', 'S', 'P', np.nan, np.nan, 'A', None],
        'age' : [20, 'NA', 18, 19, None, 'Missing', 20, None]}

DF = pd.DataFrame(DataF)
```

```python
DF
```

```python
#Missing values handling
DF.dropna() #Remove all rows with missing value
#Use inplace=True to make it permanent
```

```python
#Default synatax is for how = 'any'
DF.dropna(axis='index', how='all') #Axis = 'columns','index', 'rows'
#Use inplace=True to make it permanent
```

```python
#Based on particular columns
DF.dropna(axis='index', how='any', subset=['age', 'last'])
#Use inplace=True to make it permanent
```

```python
DF.replace('NA', np.nan, inplace=True)
DF.replace('Missing', np.nan, inplace=True)
DF
```

```python
#Check for null values
DF.isna()
```

```python
DF.fillna('MISSING')
```

```python
DF.dtypes
```

```python
DF['age'].mean()
```

```python
#Type casting of columns
DF['age'] = DF['age'].astype(object)
DF.dtypes
```
